```python
data = data.drop("Country",1)
def doCAH(X, nclust=2):
    model = AgglomerativeClustering(n_clusters=nclust, affinity = 'euclidean', linkage = 'ward')
    clust_labels1 = model.fit_predict(X)
    return (clust_labels1)

clust_labels1 = doCAH(data, 2)
CAH = pandas.DataFrame(clust_labels1)
data.insert((data.shape[1]),'CAH',CAH)
```

```python
plt.figure()
scatter = plt.scatter(data['Economy..GDP.per.Capita.'],data['Trust..Government.Corruption.'],
                     c=CAH[0],s=50)
plt.title('CAH Clustering')
plt.xlabel('Economy')
plt.ylabel('Trust')
plt.colorbar(scatter)
plt.show()
```

```python
def doAffinity(X):
    model = AffinityPropagation(damping = 0.5, max_iter = 250, affinity = 'euclidean')
    model.fit(X)
    clust_labels2 = model.predict(X)
    cent2 = model.cluster_centers_
    return (clust_labels2, cent2)

clust_labels2, cent2 = doAffinity(data)
affinity = pandas.DataFrame(clust_labels2)
data.insert((data.shape[1]),'affinity',affinity)
```

```python
plt.figure()
scatter = plt.scatter(data['Economy..GDP.per.Capita.'],data['Trust..Government.Corruption.'],
                     c=affinity[0],s=50)
plt.title('Affinity Clustering')
plt.xlabel('Economy')
plt.ylabel('Trust')
plt.colorbar(scatter)
plt.show()
```
